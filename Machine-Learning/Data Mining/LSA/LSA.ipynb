```python
from pathlib import Path
from collections import Counter
import re
import numpy as np
import pandas as pd
f = Path('titles.txt')
content = re.sub("[,':!]",'',f.read_text().lower().replace('-',' '))
terms = [line.split() for line in content.split('\n')]
series = [pd.Series(Counter(line),name=f'T{index}') for index,line in enumerate(terms,1)]
dtm = pd.concat(series,axis=1)
dtm = dtm.fillna(0)
dtm
```

```python
stopwords = ['a','the','for','of','and','to','if','abc','in','edition','little']
dtm = dtm.loc[~dtm.index.isin(stopwords)]
dtm
```

```python
keywords = ['book','dads','dummies','estate','guide','investing','market','real','rich','stock','value']
dtm = dtm.loc[keywords]
dtm
```

```python
def idf(database):
    return np.log(len(database.columns)/(database>0).sum(axis=1))
idf_dtm = idf(dtm)
idf_dtm
```

```python
def tf_idf(database):
    tf_ = database/database.sum(axis=0)
    idf_ = np.log(len(database.columns)/(database>0).sum(axis=1))
    return tf_.mul(idf_,axis=0)
tf_idf_dtm = tf_idf(dtm)
tf_idf_dtm
```

```python
def cos_similar_single(v1:np.array,v2:np.array):
    num = np.dot(v1,v2)
    denom = np.linalg.norm(v1)*np.linalg.norm(v2)
    res = num/denom
    return res
```

```python
for x in words:
    for y in words:
#         print(f'{cos_similar_single(x,y)}',end='\t')
        print("%.4f"%cos_similar_single(x,y),end='\t')
    print()
```

```python
def cos_similar_matrix(m1:np.array,m2:np.array):
    num = m1@m2.T
    denom = np.linalg.norm(m1,axis=1)[:,np.newaxis]*np.linalg.norm(m2,axis=1)
    return num/denom
cos_similar_matrix(words[1:3],words)
```

```python
df_words = pd.DataFrame(cos_similar_matrix(words,words),index=dtm.index,columns=dtm.index)
df_words
```
